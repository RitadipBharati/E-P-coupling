```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
def INT_TRAP(a,b):
    r=0
    for i in range(len(b)-1):
        h= a[i+1]-a[i]
        f_a=b[i]
        f_b=b[i+1]
        r+=h*(f_a+f_b)/2
    return(r)
hcut= 4.135667696e-15/(2*2.314159*13.605698)
```

```python
a_file=open(r"alpha2F_mote2\doubleremono\a2F.DOS1","r")
lines= a_file.readlines()
a_file.close()

new_file=open(r"alpha2F_mote2\doubleremono\a2F1.txt","w")
for i in range(len(lines)):
    if i>4 and i!=1005:
        new_file.write(lines[i])
new_file.close()

new_file=open(r"alpha2F_mote2\doubleremono\a2F1.txt","r")
Lines= new_file.readlines()
new_file.close()
k=pd.read_csv(r"alpha2F_mote2\doubleremono\a2F1.txt",sep='\s+')
k.to_csv(r"alpha2F_mote2\doubleremono\a2F1.csv",index=None)
```

```python
A2F=[r"alpha2F_mote2\doubleremono\a2F.DOS"+str(i) for i in range(1,11)]
A2f=[r"alpha2F_mote2\doubleremono\a2F"+str(i)+".txt" for i in range(1,11)]
a2f=[r"alpha2F_mote2\doubleremono\a2F"+str(i)+".csv" for i in range(1,11)]
#print(len(A2F))

for j in range(10):
    a_file=open(A2F[j],"r")
    lines= a_file.readlines()
    a_file.close()
    new_file=open(A2f[j],"w")
    for i in range(len(lines)):
        if i>4 and i!=1005:
            new_file.write(lines[i])
    new_file.close()
    k=pd.read_csv(A2f[j],sep='\s+')
    k.to_csv(a2f[j],index=None)
```

```python
a_file=open(r"mote2-new\doubleremono\lambda","r")
lines= a_file.readlines()
a_file.close()
new_file=open(r"mote2-new\doubleremono\lambda.txt","w")
for i in range(len(lines)):
    if i>1:
        new_file.write(lines[i])
new_file.close()
k=pd.read_csv(r"mote2-new\doubleremono\lambda.txt",sep='\s+')
k.to_csv(r"mote2-new\doubleremono\lambda.csv",index=None)
df=pd.read_csv(r"mote2-new\doubleremono\lambda.csv",header=None)
Lambda=df.iloc[:,(3)]
l=Lambda.to_numpy()
omegalog=df.iloc[:,(8)]
Wlog=omegalog.to_numpy()
#print(l)
Totwlog=np.sum(Wlog)
print(Totwlog)
```

```python
#CS=["Cs mag"+str(i)+".csv" for i in range(1,11)] 
#Outcs=["Cs magout"+str(i)+".csv" for i in range(1,11)]

#Lambda=[0.5215,0.4707,0.4757,0.4888,0.5079,0.5322,0.5588,0.5855,0.6112,0.6359]
#print(DF)
TotI=0
for i in range(10):
    df=pd.read_csv(a2f[i]) #Calling an element of a2f list
    #print(df)
    alpha2f = df.iloc[:,(0)] #geting 1st column(total energy)
    E= alpha2f.to_numpy() #transforming df into numpy array
    y = df.iloc[:,(1)] #getting 2nd column(total a2f)
    E1=y.to_numpy()  #transforming df into numpy array
    W=E/hcut
    I=INT_TRAP(W,E1)
    #print(I)
    w2=2*I/l[i]
    TotI+=w2
    #L=np.array([W,E1])
    #LT=L.T
    #print(LT)
    #np.savetxt(Outcs[i],LT,delimiter=",")
    plt.scatter(W,E1,marker=".")
    plt.xlabel("omega")
    plt.ylabel("a2f")
    #plt.xlim(,0.0027)
    plt.grid()
    plt.title(a2f[i])
    #plt.show()
print("Total integral="+str(TotI))  #totalintegral
#TotL=17.0059   #total lambda
#Avg_w2=2*TotI/TotL   #w2
w2=TotI**0.5      #sqrt(w2)
print("square root of w2="+str(w2))
ratio=Totwlog/w2
print("Total wlog/square root of w2="+str(ratio))
print("Total wlog="+str(Totwlog))
```

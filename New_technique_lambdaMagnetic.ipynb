```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
def INT_TRAP(a,b):
    r=0
    for i in range(len(b)-1):
        h= a[i+1]-a[i]
        f_a=b[i]
        f_b=b[i+1]
        r+=h*(f_a+f_b)/2
    return(r)
hcut= 4.135667696e-15/(2*2.314159*13.605698)
```

```python
df1 = pd.read_csv("Cs mag1.csv")  
alpha2f = df1.iloc[:,(0)]
#print(X)
E= alpha2f.to_numpy()
#<w>=2/lambda(int(alpha2f*wdw))
y = df1.iloc[:,(1)]
#print(y)
E1=y.to_numpy()
W=E1/hcut
I1=INT_TRAP(W,E)
print(I1)
```

```python
CS=["Cs mag"+str(i)+".csv" for i in range(1,11)] 
Outcs=["Cs magout"+str(i)+".csv" for i in range(1,11)]
Lambda=[2.3351,1.7725,1.4529,1.2660,1.1348,1.0384,0.9661,0.9119,0.8720,0.8433]
#print(DF)
TotI=0
for i in range(10):
    df=pd.read_csv(CS[i]) #Calling an element of CS list
    #print(df)
    alpha2f = df.iloc[:,(0)] #geting 1st column(total energy)
    E= alpha2f.to_numpy() #transforming df into numpy array
    y = df.iloc[:,(1)] #getting 2nd column(total a2f)
    E1=y.to_numpy()  #transforming df into numpy array
    W=E/hcut
    I=INT_TRAP(W,E1)
    #print(I)
    w2=2*I/Lambda[i]
    TotI+=w2
    L=np.array([W,E1])
    LT=L.T
    #print(LT)
    #np.savetxt(Outcs[i],LT,delimiter=",")
    plt.scatter(W,E1,marker=".")
    plt.xlabel("omega")
    plt.ylabel("a2f")
    #plt.xlim(,0.0027)
    plt.grid()
    plt.title(CS[i])
    #plt.show()
print(TotI)  #totalintegral
#TotL=17.0059   #total lambda
#Avg_w2=2*TotI/TotL   #w2
w2=TotI**0.5      #sqrt(w2)
print(w2)
Totwlog=823.3201
Csratio=Totwlog/w2
print(Csratio)
```
